```python
import pandas as pd

def create_telefones_df(lista_telefones):
    df = pd.DataFrame(data=lista_telefones, columns=['telefone'])
    df = df.dropna()
    df['telefone'] = df['telefone'].astype(str)
    df['telefone'] = df['telefone'].loc[(df['telefone'].str.len()==10) | (df['telefone'].str.len()==11)]

    if df.empty:
        return []
        
    df['ddd'] = df['telefone'].str[:2]
    df['numero'] = df['telefone'].str[2:]
    df['id_resultado'] = 1
    df.at[0, 'id_resultado'] = 8
    
    df = df[['ddd', 'numero', 'id_resultado']]
  
    return df.to_dict('records')


def extract_emails_df(lista_emails):
    df = pd.DataFrame(data=lista_emails, columns=['email'])
    df = df.dropna()
    
    if df.empty:
        return []
    
    df['email'] = df['email'].astype(str)
    df['id_status'] = 1
    
    df = df[['email', 'id_status']]
    
    return df.to_dict('records')


def extract_endereco_df(lista_endereco):
    df = pd.DataFrame(data=lista_endereco, columns=['endereco'])
    df = df.dropna()
    
    if df.empty:
        return []
    
    df[['logradouro', 'numero', 'complemento', 'bairro', 'cidade', 'uf', 'cep']] = df['endereco'].str.split(",",expand=True)
    
    cols = ['logradouro', 'numero', 'complemento', 'bairro', 'cidade', 'uf', 'cep']

    for col in cols:
        df[col] = df[col].apply(lambda x: x if x else None)
    
    df['id_status'] = 1
    df.drop(columns=['endereco'], inplace=True)   
    return df.to_dict('records')


def fill_zeros(x):
    if (len(x) > 11 and len(x) < 14):
        return x.zfill(14)
    else:
        return x
    

df = pd.read_csv(r'file.txt', sep=';', dtype='str', encoding='latin-1')
df['id_origem'] = #insert id_origem

df['TELEFONE'] =  df.apply(lambda x: list([x['TELEFONE1'],
                                           x['TELEFONE2'],
                                           x['TELEFONE3'],
                                           x['TELEFONE4'],
                                           x['TELEFONE5'],
                                           x['TELEFONE6'],
                                           x['TELEFONE7'],
                                           x['TELEFONE8'],
                                           x['TELEFONE9'],
                                           x['TELEFONE10']]),axis=1)


df['ligacoes'] = df['TELEFONE'].apply(lambda x: create_telefones_df(x))

df = df[['NR_CPF_CNPJ2', 'DS_NOME', 'CD_UF', 'DS_EMAIL', 'EXTRA2', 'EXTRA4', 'id_origem', 'ligacoes']]

# EMAILS
# cria coluna com os valores da ds_email como uma lista 
df['DS_EMAIL'] =  df.apply(lambda x: list([x['DS_EMAIL']]), axis=1)

df['emails'] = df['DS_EMAIL'].apply(lambda x: extract_emails_df(x))

df = df[['NR_CPF_CNPJ2', 'DS_NOME', 'EXTRA2', 'EXTRA4', 'id_origem', 'ligacoes', 'emails']]

# cria coluna com os valores da EXTRA4 como uma lista 
df['EXTRA4'] =  df.apply(lambda x: list([x['EXTRA4']]), axis=1)

df['enderecos'] = df['EXTRA4'].apply(lambda x: extract_endereco_df(x))

df.to_dict('records')

df.rename(columns={'DS_NOME': 'nome', 'NR_CPF_CNPJ2': 'cpf_cnpj'}, inplace=True)
df = df[['id_origem', 'nome', 'cpf_cnpj', 'EXTRA2', 'emails', 'enderecos', 'ligacoes']]

df.to_dict('records')
df.head(5)

df.loc[df['EXTRA2'].notna(), 'id_status_digital_cliente'] = '2'
df.loc[df['EXTRA2'].isna(), 'id_status_digital_cliente'] = '1'

df = df[(df['cpf_cnpj'].str.len() >= 11) & (df['cpf_cnpj'].str.len() <= 14)]

df['cpf_cnpj'] = df['cpf_cnpj'].apply(lambda x: fill_zeros(x))

df = df[['id_origem', 'nome', 'cpf_cnpj', 'id_status_digital_cliente', 'emails', 'enderecos', 'ligacoes',]]


df.to_dict('records')
```
